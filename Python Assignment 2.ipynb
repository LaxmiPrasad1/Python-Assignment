## Question 1

Printing the Pattern

```python
k=1
for i in range(0,9):
    if i>=5:
        for j in range(0,i-k):
            print('*',end='')
        k+=2
    else:
        for j in range(0,i+1):
            print('*',end='')
    print('\n')
```

## Question 2 
Write a Python program to reverse a word after accepting the input from the user.

```python
name=input("Enter the String: ")
name=name[::-1]
print(name)
```
